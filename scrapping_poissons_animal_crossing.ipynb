```python
from urllib.request import Request, urlopen
import pandas as pd
from bs4 import BeautifulSoup


url_base = "https://www.millenium.org/guide/361539.html"

req = Request(url_base, headers={'User-Agent': 'Mozilla/5.0'})

webpage = urlopen(req).read()
```

```python
soup = BeautifulSoup(webpage, "html.parser")
article = soup.find_all(class_="w-list-items__item")
```

```python
print(article)
article = [str(i) for i in article]
```

```python
article
```

```python
#On récupère le nom du poisson, et sa rareté.
scrappable_html = article[2].split("\n")
for i in range(len(scrappable_html)):
    if "img" in scrappable_html[i]:
        link_img=scrappable_html[i].split(r'"')[3]
    if "<" not in scrappable_html[i]:
        nom_poisson = scrappable_html[i]
    if "Rareté" in scrappable_html[i]:
        rarete = scrappable_html[i].split("Rareté</strong> :")[1].replace("</p>","")

print(link_img,nom_poisson,rarete)


    
```

```python
import pandas as pd

df = pd.DataFrame(columns=["nom_poisson","rarete","link_img"])

for j in range(len(article)):
    scrappable_html = article[j].split("\n")
    for i in range(len(scrappable_html)):
        if "img" in scrappable_html[i]:
            link_img=scrappable_html[i].split(r'"')[3]
        if "<" not in scrappable_html[i] and ">" not in scrappable_html[i] and scrappable_html[i] != "":
            nom_poisson = scrappable_html[i]
            print(nom_poisson,i)
        if "Rareté" in scrappable_html[i]:
            rarete = scrappable_html[i].split("Rareté</strong> :")[1].replace("</p>","") 
    nouvelle_ligne = {"nom_poisson" : nom_poisson , "rarete" : rarete , "link_img" : link_img }
    df= pd.concat([df, pd.DataFrame([nouvelle_ligne])],ignore_index=True)
```
